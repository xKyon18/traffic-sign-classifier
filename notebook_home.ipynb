```python
import os
import pickle
import numpy as np
from PIL import Image
```

```python
baseDirectory = os.path.join(os.getcwd(), 'Train_Images')
trainData = []
print(baseDirectory)
print(trainData)
```

```python
for folder in os.listdir(baseDirectory):
    folderPath = os.path.join(baseDirectory, folder)
    files = []
    for filename in os.listdir(folderPath):
        if filename.endswith('ppm'):
            filePath = os.path.join(folderPath, filename)
            with Image.open(filePath) as img:
                fileContent = np.array(img)
                files.append(fileContent)
    trainData.append(files)

```

```python
import matplotlib.pyplot as plt

print(len(trainData))

res = []
for i in range(len(trainData)):
    res.append(len(trainData[i]))
print(res)
```

```python
import random

plt.bar([i for i in range(43)], res)
plt.title('Traffic Sign Classification Data')
plt.ylabel('No. of Examples')
plt.xlabel('Categories')
plt.show()

for i, cat in enumerate(trainData):
    print(f'Class {i + 1}: {len(trainData[i])} samples')
    fig = plt.figure(figsize=(6, 1))
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)
    for j in range(10):
        axis = fig.add_subplot(1, 10, j + 1, xticks=[], yticks=[])
        axis.imshow(cat[random.randint(0, len(cat) - 1)])
    plt.show()


        
```

```python
with open('trainData.pkl', 'wb') as file:
    pickle.dump(trainData, file)
```

```python
with open('trainData.pkl', 'rb') as file:
    loadedData = pickle.load(file)

print(len(loadedData))
for cat in loadedData:
    print(len(cat))
```
